```python
import ipyplot
from rich import print

from pyinaturalist import (
    enable_logging,
    iNatClient,
    pprint,
)

enable_logging()
client = iNatClient()
```

## Basic observation search
We'll start with the observation search page. [client.observations.search()](https://pyinaturalist.readthedocs.io/en/stable/modules/pyinaturalist.controllers.ObservationController.html#pyinaturalist.controllers.ObservationController.search)
supports all the same search filters as the ones you see on iNaturalist.org:

![inat-observation-search-bar.png](images/inat-observation-search-bar.png)

![inat-observation-search-filters.png](images/inat-observation-search-filters.png)

Let's start by searching for all of your own observations:

```python
# Replace with your own username
USERNAME = 'jkcook'

my_observations = client.observations.search(user_id=USERNAME).all()
```

## Observation data
Take a look at one of the observations to see what information it contains:

```python
print(my_observations[0])
```

### Observation data compared to the web UI
Here is how some of those fields correspond to what you see on an observation page on iNaturalist.org:
<img src="https://github.com/pyinat/pyinaturalist/raw/main/docs/images/inat-observation-page-annotated.png" />

You'll notice that there are many more fields available; see the [Observation docs](https://pyinaturalist.readthedocs.io/en/stable/modules/pyinaturalist.models.Observation.html) for a complete list.

### Previewing data
In many cases, you will want to quickly preview API results without looking through the full details for each result. `pyinaturalist.pprint()` can be used to show a condensed table of almost all response types. Here's an example with just the first 30 results:

```python
pprint(my_observations[:30])
```

## Observation species
On iNaturalist.org, the next tab of the observation search page is **Species**.
You can get this information with [client.observations.species_counts()](https://pyinaturalist.readthedocs.io/en/stable/modules/pyinaturalist.controllers.ObservationController.html#pyinaturalist.controllers.ObservationController.species_counts).


For example, [all the frogs and toads observed in Mexico](https://www.inaturalist.org/observations?place_id=6793&taxon_id=20979&view=species):

![inat-observation-species-counts.png](images/inat-observation-species-counts.png)

Here is how to get that same information from the API:

```python
# Note: 6793 is the place ID for Mexico, and 20979 is the taxon ID for the Order Anura
taxa = client.observations.species_counts(place_id=6793, taxon_id=20979)
```

Then we can see a preview of the top 10 results:

```python
pprint(taxa[:20])
```

## Observation identifiers
[client.observations.identifiers()](https://pyinaturalist.readthedocs.io/en/stable/modules/pyinaturalist.controllers.ObservationController.html#pyinaturalist.controllers.ObservationController.identifiers) gets us information shown on the next tab: **Identifiers**:

![inat-observation-identifiers.png](images/inat-observation-identifiers.png)

```python
users = client.observations.identifiers(place_id=6793, taxon_id=20979)
pprint(users[:10])
```

## Observation observers
And [client.observations.observers()](https://pyinaturalist.readthedocs.io/en/stable/modules/pyinaturalist.controllers.ObservationController.html#pyinaturalist.controllers.ObservationController.observers) gets us information from the **Observers** tab:

![inat-observation-observers.png](images/inat-observation-observers.png)

```python
users = client.observations.observers(place_id=6793, taxon_id=20979)
pprint(users[:10])
```

## Observation photos
When you're working in Jupyter, there are a number of ways to preview observation photos.
For these examples, we'll use your own observation data from the first step in this tutorial.

### Viewing individual observation photos
Use `Photo.show()` to see a photo from a single observation:

```python
my_observations[-4].photos[0].show()
```

### Observation photo grid
We can use [ipyplot](https://github.com/karolzak/ipyplot) to preview observation images. `Observation.photos` contains a list of [Photo](https://pyinaturalist.readthedocs.io/en/stable/modules/pyinaturalist.models.Photo.html#pyinaturalist.models.Photo) objects, and we can use those to get a thumnail URL for first photo from each observation. For image labels, just call `str(observation)` to get a summary of the observation (who/what/when/where).

```python
images = [obs.photos[0].thumbnail_url for obs in my_observations[:15]]
labels = [obs.taxon.full_name for obs in my_observations[:15]]
ipyplot.plot_images(images, labels)
```

### Observation photo grid grouped by iconic taxon
We can organize this a bit more by grouping these photos by iconic taxon. Use `ipyplot.plot_class_tabs` to group by label, and use `Observation.taxon.iconic_taxon_name` as the image labels:

```python
images = [obs.photos[0].thumbnail_url for obs in my_observations]
labels = [obs.taxon.iconic_taxon_name for obs in my_observations]
ipyplot.plot_class_tabs(images, labels, max_imgs_per_tab=15)
```

## Observation histogram
Another useful format is the
[observation histogram](https://pyinaturalist.readthedocs.io/en/stable/modules/pyinaturalist.controllers.ObservationController.html#pyinaturalist.controllers.ObservationController.histogram),
which shows the number of observations over a given `interval`.

The default is `month_of_year`, which will show counts of all your observations by month, for all years combined:

```python
histogram = client.observations.histogram(user_id=USERNAME)
pprint(histogram)
```

The underlying data is just a dict of `{month: count}`:

```python
pprint(histogram.raw)
```

Or use `month` + a date range to get monthly counts for a single year:

```python
histogram = client.observations.histogram(
    user_id=USERNAME, interval='month', d1='2020-01-01', d2='2020-12-31'
)
pprint(histogram)
```
